# Solutions discrete Fourier transform

```python
%matplotlib inline
import numpy as np
from numpy.fft import fft, ifft
import matplotlib.pyplot as plt

N = 128
x = np.linspace(0, 2*np.pi, N, endpoint=False)
fx = np.sin(10*x)
ftf = fft(fx)

fig, axes = plt.subplots(2,2)
axes[0,0].plot(x, fx)
axes[0,0].set_ylabel("f(x)")
axes[0,1].semilogy(abs(ftf))
#axes[0,1].set_ylim(10**-3, 10)
axes[0,1].set_ylabel("|f(k)|")
#axes[0,1].set_xlim(0,20)
axes[1,0].plot(ftf.real)
axes[1,0].set_ylabel("Real(f(k))")
axes[1,1].plot(ftf.imag)
axes[1,1].set_ylabel("Imag(f(k))")
plt.show()
```

```python
fx = np.sin(x)
ftf = fft(fx)
fx2 = ifft(np.exp(1j*np.pi/2)*ftf)

fig, axes = plt.subplots(2,2)
axes[0,0].plot(x, fx)
axes[0,0].set_ylabel("f(x)")
axes[0,1].plot(x, abs(fx2))
axes[0,1].set_ylabel("|f_2(x)|")
axes[1,0].plot(x, fx2.real)
axes[1,0].set_ylabel("Real(f_2(x))")
axes[1,1].plot(x, fx2.imag)
axes[1,1].set_ylabel("Imag(f_2(x))")
plt.show()
```

```python
fx = np.sin(x)
ftf = fft(fx)
ftf[:N//2] *= np.exp(1j*np.pi/2)
ftf[N//2:] *= np.exp(-1j*np.pi/2)
fx2 = ifft(ftf)

fig, axes = plt.subplots(2,2)
axes[0,0].plot(x, fx)
axes[0,0].set_ylabel("f(x)")
axes[0,1].plot(x, abs(fx2))
axes[0,1].set_ylabel("|f_2(x)|")
axes[1,0].plot(x, fx2.real)
axes[1,0].set_ylabel("Real(f_2(x))")
axes[1,1].plot(x, fx2.imag)
axes[1,1].set_ylabel("Imag(f_2(x))")
plt.show()
```
