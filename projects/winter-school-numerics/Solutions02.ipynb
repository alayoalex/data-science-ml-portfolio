```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi/2, 4)
f = np.sin

xplot = np.linspace(-.2, np.pi/2+.2, 200)
plt.plot(xplot, f(xplot), color='k')

fx = np.concatenate((f(x),[0]))
xp = np.concatenate((x,[x[-1]]))
plt.fill(xp, fx)
plt.show()
```

```python
def trapez(f, a, b, h):
    N = int(np.ceil((b-a)/h))+1
    print(f'N={N}')
    xn = np.linspace(a, b, N)
    return (b-a)/(N-1)*(np.sum(f(xn))-.5*(f(a)+f(b)))


analytical = 1.0
print(trapez(f, 0, np.pi/2, .001)-analytical)
```

```python
def trapez_nonconst(f, xn):
    return .5*np.sum((xn[1:]-xn[:-1])*(f(xn[1:]) + f(xn[:-1])))

analytical = 1.0
N = int(np.ceil((np.pi/2)/0.001))+1
xn = np.linspace(0, np.pi/2, N)
print(trapez_nonconst(f, xn)-analytical)
```

```python
from scipy.integrate import quad

integral, estimated_err = quad(np.sin, 0, np.pi/2) 
print(integral-analytical, estimated_err)
```
